```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('../data/chapter_9.1.csv')
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
X3=df[df['product_id']=="C"]
plt.plot(X3['expo_time'],X3['decay_rate'],color='b')
plt.show()
```

```python
#显示曝光次数与用户激活衰退率关系
plt.figure(figsize=(10,8))
plt.rcParams['font.sans-serif']=['SimHei']
X=df['expo_time']
y=df['decay_rate']
plt.scatter(X,y,color='blue')
plt.xlabel('active_times')
plt.ylabel(' decay_rate ')
# plt.xlim(0,1.4)
# plt.ylim(0,1)
plt.show()
```

```python

#构建关系模型
for i in range(0,25):
    j=list(set(df['product_id']))[i]
    df1=df[df['product_id']==j]
    X=df1['expo_time']
    y=df1['decay_rate']
    lr=LinearRegression()
    lr.fit(np.array(X).reshape(-1,1),y)
    y_pred=lr.predict(np.array(X).reshape(-1,1))
    r2=r2_score(y,y_pred)
    print(r2)
```
